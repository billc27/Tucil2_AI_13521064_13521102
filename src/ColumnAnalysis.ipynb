```python
# Import library pandas
import pandas as pd

# Read csv file
df = pd.read_csv("../data/data_train.csv")

import pandas as pd
import matplotlib.pyplot as plt

# For numeric
def analyze_numeric_distribution(df, column):
    data = df[column]
    
    # Data distribution plot
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=25, color='orange')
    plt.title(f'Distribusi Data untuk Kolom {column}')
    plt.xlabel(column)
    plt.ylabel('Frekuensi')
    plt.grid(True)
    plt.show()
    
    # Curtosis Analysis
    kurt = df[column].kurtosis()
    print(f'Kurtosis untuk kolom {column}: {kurt}')
    print("=========================================")
    print("\n")

# For non-numeric
def plot_categorical_histogram(df, column):
    data = df[column]
    
    # Plot histogram
    plt.figure(figsize=(10, 6))
    data.value_counts().plot(kind='bar', color='cyan')
    plt.title(f'Histogram untuk Kolom {column}')
    plt.xlabel(column)
    plt.ylabel('Frekuensi')
    plt.grid(True)
    plt.show()
    print("=========================================")
    print("\n")

# Analyze all columns
def analyze_all_columns(df):
    non_numerical_columns = ['blue', 'dual_sim', 'four_g', 'three_g', 'touchscreen', 'wifi']

    for column in df.columns:
        print("====================o====================")

        if column not in non_numerical_columns: # if numerical
            print(f"Data Distributon - Column {column}")
            analyze_numeric_distribution(df, column)
        else: # if not numerical
            print(f"Histogram - Column {column}")
            plot_categorical_histogram(df, column)

        print("====================o====================")
```

COLUMNS ANALYSIS

```python
analyze_all_columns(df)
```
