## 1. Data Train Statistics

```python
# Import pandas
import pandas as pd

# Read csv file
df = pd.read_csv("../data/data_train.csv")

# Print basic statistics function
def print_descriptive_statistics(dataframe, column_name):
    print("====================o====================")
    print(f"Statistika Dasar Kolom {column_name}")
    print("====================o====================")

    # Mean
    print("Mean:", dataframe[column_name].mean())
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # Median
    print("Median:", dataframe[column_name].median())
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # Modus
    print("Modus:") 
    all_modes = dataframe[column_name].mode().values.tolist()
    if (len(all_modes) == dataframe[column_name].count()):
        print("Ada", dataframe[column_name].count(), "modus pada kolom ini. Jumlah tersebut sama dengan jumlah nilai pada kolom ini.")
        print("Hal tersebut menandakan bahwa kolom ini memiliki nilai-nilai yang berbeda satu sama lain.")
    else:
        for mode in all_modes:
            print(mode)
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # Standar Deviasi
    print("Standar Deviasi:", dataframe[column_name].std())
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # Variansi
    print("Variansi:", dataframe[column_name].var())
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # Range
    print("Range:", dataframe[column_name].max() - dataframe[column_name].min())
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # Minimum
    print("Nilai Minimum:", dataframe[column_name].min())
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # Maximum
    print("Nilai Maksimum:", dataframe[column_name].max())
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # Kuartil
    print("Kuartil Bawah:", dataframe[column_name].quantile(0.25))
    print("Kuartil Tengah:", dataframe[column_name].quantile(0.50))
    print("Kuartil Atas:", dataframe[column_name].quantile(0.75))
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # IQR
    print("IQR:", dataframe[column_name].quantile(0.75) - dataframe[column_name].quantile(0.25))
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # Skewness
    print("Skewness:", dataframe[column_name].skew())
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")

    # Kurtosis
    print("Kurtosis:", dataframe[column_name].kurtosis())
    print("=========================================")
    print("\n")
```

```python
# Print out statistics for each column
for column in df.columns:
    print_descriptive_statistics(df, column)
```
