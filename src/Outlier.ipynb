```python
# Import library pandas
import pandas as pd

# Read csv file
df = pd.read_csv("../data/data_train.csv")


def print_outliers(dataframe, column_name):
    print("====================o====================")
    print(f"Outliers Column {column_name}")
    print("====================o====================")

    # Find outliers in a specific column
    Q1 = dataframe[column_name].quantile(0.25)
    Q3 = dataframe[column_name].quantile(0.75)
    IQR = Q3 - Q1

    outliers = dataframe[(dataframe[column_name] < (Q1 - 1.5 * IQR)) | (dataframe[column_name] > (Q3 + 1.5 * IQR))]
    print(f"Outliers in {column_name}:")
    print(outliers)
    print("\n")
```

OUTLIER

```python
for column in df.columns:
    print_outliers(df, column)
```
