```python
# Import library pandas
import pandas as pd

# Read csv file
df = pd.read_csv("../data/data_train.csv")

def print_missing_values(dataframe, column_name):
    print("====================o====================")
    print(f"Missing Values Column {column_name}")
    print("====================o====================")

    # Find missing values in a specific column
    missing_values = dataframe[column_name].isnull().sum()
    print(f"Missing Values in {column_name}:")
    print(missing_values)
    print("\n")
```

MISSING VALUE

```python
for column in df.columns:
    print_missing_values(df, column)
```
